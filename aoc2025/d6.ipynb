```python
from aocd import get_puzzle

puzzle = get_puzzle(year=2025, day=6)
data = puzzle.input_data
examples = puzzle.examples
```

## Part 1

```python
from math import prod

problems = list(zip(*[l.split() for l in data.split('\n')]))
problems = [
    [p[-1], [int(a) for a in p[:-1]]] for p in problems
]

total = 0

for p in problems:
    match p:
        case ['+', nums]:
            total += sum(nums)
        case ['*', nums]:
            total += prod(nums)

print(total)
```

## Part 2

```python
from math import prod

unflippedproblems = zip(*data.split('\n'))
problems = []

for col in unflippedproblems:
    if col[-1] != ' ':
        if problems:
            problems[-1][1] = list(reversed(problems[-1][1]))
        problems.append([col[-1], []])

    try:
        problems[-1][1].append(int(''.join(col[:-1])))
    except:
        pass

total = 0

for p in problems:
    match p:
        case ['+', nums]:
            total += sum(nums)
        case ['*', nums]:
            total += prod(nums)

print(total)
```
