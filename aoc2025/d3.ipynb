```python
from aocd import get_puzzle

puzzle = get_puzzle(year=2025, day=3)
data = puzzle.input_data
examples = puzzle.examples
```

## Part 1

```python
joltage = 0

for bank in data.split('\n'):
    jolt = ['0', '0']
    for battery in bank[:-1]:
        if battery > jolt[0]:
            jolt[0] = battery
            jolt[1] = "0"
        elif battery > jolt[1]:
            jolt[1] = battery

    if bank[-1] > jolt[1]:
        jolt[1] = bank[-1]

    joltage += int(''.join(jolt))

print(joltage)
```

## Part 2

```python
joltage = 0

for bank in data.split('\n'):
    jolt = ['0'] * 12
    
    for i, battery in enumerate(bank):
        for j in range(12):
            if i > (len(bank) - 12 + j):
                continue
            if battery > jolt[j]:
                jolt[j] = battery
                jolt[j+1:] = ['0' for _ in jolt[j+1:]]
                break

    joltage += int(''.join(jolt))

print(joltage)
```
