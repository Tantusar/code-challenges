```python
from aocd import get_puzzle

puzzle = get_puzzle(year=2025, day=5)
data = puzzle.input_data
examples = puzzle.examples
```

## Part 1

```python
ranges, ingredients = data.split('\n\n')

ranges = sorted([
    [int(a) for a in r.split('-')] for r in ranges.split('\n')
])

ingredients = sorted([int(i) for i in ingredients.split('\n')])
```

```python
fresh = 0
pointer = 0
done = False

for i in ingredients:
    if i < ranges[pointer][0]:
        continue
    
    while i >= ranges[pointer][0]:
        if i <= ranges[pointer][1]:
            fresh += 1
            break
        pointer += 1
        if pointer >= len(ranges):
            done = True
            break
        
    if done: break

print(fresh)
```

## Part 2

```python
collapsedranges = ranges.copy()
pointer = 0

while pointer < len(collapsedranges) - 1:
    if collapsedranges[pointer][1] >= collapsedranges[pointer+1][0]:
        collapsedranges[pointer:pointer+2] = [
            [
                collapsedranges[pointer][0], max(
                    collapsedranges[pointer][1],
                    collapsedranges[pointer+1][1]
                )
            ]
        ]
    else:
        pointer += 1

print(sum([r[1] - r[0] + 1 for r in collapsedranges]))
```
