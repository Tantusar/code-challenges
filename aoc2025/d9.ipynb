```python
from aocd import get_puzzle

puzzle = get_puzzle(year=2025, day=9)
data = puzzle.input_data
examples = puzzle.examples
```

## Part 1

```python
from itertools import combinations

tiles = [[int(a) for a in d.split(',')] for d in data.split('\n')]

tilesizes = [
    [(abs(a[0]-b[0]) + 1) * (abs(a[1]-b[1]) + 1), a, b] for a, b in combinations(tiles, 2)
]

print(max(tilesizes)[0])

p1box = shapely.box(*max(tilesizes)[1], *max(tilesizes)[2])
```

## Part 2

```python
from itertools import combinations
import shapely

tiles = [[int(a) for a in d.split(',')] for d in data.split('\n')]

polygon = shapely.Polygon(tiles + [tiles[0]])

tilesizes = []

for a, b in combinations(tiles, 2):
    if shapely.contains(polygon, shapely.box(*a, *b)):
        tilesizes.append([(abs(a[0]-b[0]) + 1) * (abs(a[1]-b[1]) + 1), a, b])

print(max(tilesizes)[0])

p2box = shapely.box(*max(tilesizes)[1], *max(tilesizes)[2])
```

```python
from shapely import plotting
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (16, 16))

plotting.plot_polygon(polygon, ax)
plotting.plot_polygon(p1box, ax, color='green')
plotting.plot_polygon(p2box, ax, color='red')
```
