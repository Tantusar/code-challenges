```python
from aocd import get_puzzle

puzzle = get_puzzle(year=2025, day=1)
data = puzzle.input_data
examples = puzzle.examples
```

## Part 1

```python
dial = 50
password = 0

for instruction in data.split('\n'):
    match (instruction[0], instruction[1:]):
        case ('L', i):
            dial -= int(i)
        case ('R', i):
            dial += int(i)
    
    dial %= 100
    if dial == 0:
        password += 1

print(password)
```

## Part 2

```python
dial = 50
password = 0

for instruction in data.split('\n'):
    safe = (dial != 0)
    match (instruction[0], instruction[1:]):
        case ('L', i):
            dial -= int(i)
        case ('R', i):
            dial += int(i)

    if dial <= 0:
        password += (abs(dial) + (100 if safe else 0)) // 100
    elif dial >= 100:
        password += dial // 100
    
    dial %= 100

print(password)
```
