```python
from aocd import get_puzzle

puzzle = get_puzzle(year=2025, day=4)
data = puzzle.input_data
examples = puzzle.examples
```

## Part 1

```python
from itertools import product

rolls = set()

for x, i in enumerate(data.split('\n')):
    for y, j in enumerate(i):
        if j == '@':
            rolls.add((x, y))

good = 0

for x, y in rolls:
    working = set((x+a, y+b) for a, b in product([-1, 0, 1], repeat=2) if not (a == b == 0))

    if len(rolls.intersection(working)) < 4:
        good += 1

print(good)
```

## Part 2

```python
from itertools import product

rolls = set()

for x, i in enumerate(data.split('\n')):
    for y, j in enumerate(i):
        if j == '@':
            rolls.add((x, y))

good = 0
while True:
    removal = set()

    for x, y in rolls:
        working = set((x+a, y+b) for a, b in product([-1, 0, 1], repeat=2) if not (a == b == 0))

        if len(rolls.intersection(working)) < 4:
            good += 1
            removal.add((x, y))

    if not removal:
        break

    rolls.difference_update(removal)

print(good)
```

```python
from itertools import product

rolls = set()

for x, i in enumerate(data.split('\n')):
    for y, j in enumerate(i):
        if j == '@':
            rolls.add((x, y))

good = 0
checking = set(rolls)
while True:
    removal = set()
    nextchecking = set()

    for x, y in checking:
        working = set((x+a, y+b) for a, b in product([-1, 0, 1], repeat=2))

        if len(rolls.intersection(working)) <= 4:
            good += 1
            removal.add((x, y))
            nextchecking.update(rolls.intersection(working))


    if not removal:
        break

    rolls.difference_update(removal)
    checking = nextchecking.intersection(rolls)

print(good)
```
