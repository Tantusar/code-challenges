```python
from aocd import get_puzzle

puzzle = get_puzzle(year=2025, day=12)
data = puzzle.input_data
examples = puzzle.examples
```

## Part 1

```python
import numpy as np

*presents, trees = data.split('\n\n')

presents = [p.split('\n')[1:] for p in presents]
presents = [
    np.array([[1 if c == '#' else 0 for c in l] for l in p]) for p in presents
]

parsedtrees = []
for t in trees.split('\n'):
    dims, _, reqs = t.partition(': ')
    reqs = [int(r) for r in reqs.split()]
    dims = [int(d) for d in dims.split('x')]

    parsedtrees.append({
        "dims": dims,
        "reqs": reqs
    })

trees = parsedtrees
del parsedtrees
```

```python
from math import prod

presentsizes = [np.sum(p) for p in presents]

total = 0

# The naive approach just works.
for t in trees:
    dims = prod(t['dims'])
    reqs = 0
    for r in zip(presentsizes, t['reqs']):
        reqs += prod(r)
    if reqs <= dims:
        total += 1

print(total)
```
