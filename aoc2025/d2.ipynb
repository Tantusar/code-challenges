```python
from aocd import get_puzzle

puzzle = get_puzzle(year=2025, day=2)
data = puzzle.input_data
examples = puzzle.examples
```

## Part 1

```python
idtotal = 0

for r in data.split(','):
    bottom, top = r.split('-')
    bottom, top = int(bottom), int(top) + 1

    for i in range(bottom, top):
        working = str(i)
        if len(working) % 2 != 0:
            continue
        
        if working[:len(working)//2] == working[len(working)//2:]:
            idtotal += i
            # print(i)

print(idtotal)
```

## Part 2

```python
from itertools import batched

idtotal = 0

for r in data.split(','):
    bottom, top = r.split('-')
    bottom, top = int(bottom), int(top) + 1

    for i in range(bottom, top):
        working = str(i)

        for l in range(1, len(working)):
            if len(working) % l != 0:
                continue
            
            if len(set(batched(working, l))) == 1:
                idtotal += i
                break
                # print(i)

print(idtotal)
```
