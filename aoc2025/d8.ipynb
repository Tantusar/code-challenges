```python
from aocd import get_puzzle

puzzle = get_puzzle(year=2025, day=8)
data = puzzle.input_data
examples = puzzle.examples
```

## Part 1

```python
from collections import Counter
from itertools import combinations
from math import dist, prod
import networkx as nx

junctions = [tuple([int(a) for a in l.split(',')]) for l in data.split('\n')]
connections = Counter()

for a, b in combinations(junctions, 2):
    connections[(a, b)] = -dist(a, b)

connections = [c[0] for c in connections.most_common(1000)]

G = nx.Graph()

G.add_edges_from(connections)

sizes = [len(g) for g in sorted(nx.connected_components(G), key=len, reverse=True)]

print(prod(sizes[:3]))
```

## Part 2

```python
from collections import Counter
from itertools import combinations
from math import dist
import networkx as nx

junctions = [tuple([int(a) for a in l.split(',')]) for l in data.split('\n')]
connections = Counter()

for a, b in combinations(junctions, 2):
    connections[(a, b)] = -dist(a, b)

G = nx.Graph()

for c in connections.most_common():
    G.add_edge(*c[0])
    if G.number_of_nodes() == len(junctions) and nx.is_connected(G):
        print(c[0][0][0] * c[0][1][0])
        break
```
