```python
from ecd import get_inputs

event, quest = 2025, 8
```

## Part 1

```python
data1 = get_inputs(event=event, quest=quest)['1']

data1 = [int(n) for n in data1.split(',')]
```

```python
from itertools import pairwise

print([a % 16 == b % 16 for a, b in pairwise(data1)].count(True))
```

## Part 2

```python
data2 = get_inputs(event=event, quest=quest)['2']

data2 = [int(n) for n in data2.split(',')]
```

```python
from itertools import pairwise

art2 = set()

knots2 = 0

for a, b in pairwise(data2):
    if a > b:
        b, a = a, b

    for c, d in art2:
        if (c < a < d and (b < c or d < b)) or (c < b < d and (a < c or d < a)):
            knots2 += 1

    art2.add((a, b))

print(knots2)
```

## Part 3

```python
data3 = get_inputs(event=event, quest=quest)['3']

data3 = [int(n) for n in data3.split(',')]
```

```python
from itertools import pairwise, combinations

art3 = []

for c, d in pairwise(data3):
    if c > d:
        d, c = c, d
    art3.append((c, d))

result = 0

for a, b in combinations(range(1,257), 2):
    if b == 256:
        print(a, end='\r')
    working = 0

    for c, d in art3:
        if (c < a < d and (b < c or d < b)) or (c < b < d and (a < c or d < a)) or (a, b) == (c, d):
            working += 1

    result = max(result, working)

print(f"\n{result}")
```
