```python
def fishboner(data):
    fishbone = [[None, data[0], None]]

    for d in data[1:]:
        for x in range(len(fishbone)):
            if d < fishbone[x][1] and fishbone[x][0] == None:
                fishbone[x][0] = d
                break
            elif d > fishbone[x][1] and fishbone[x][2] == None:
                fishbone[x][2] = d
                break
        else:
            fishbone.append([None, d, None])

    return int(''.join(str(f[1]) for f in fishbone))
```

## Part 3

```python
data3 = get_inputs(event=event, quest=quest)['3']

data3dict = {}

for l in data3.split('\n'):
    ident, _, dt = l.partition(':')

    data3dict[int(ident)] = [int(n) for n in dt.split(',')]
```

```python
def fishboner2(data):
    fishbone = [[None, data[0], None]]

    for d in data[1:]:
        for x in range(len(fishbone)):
            if d < fishbone[x][1] and fishbone[x][0] == None:
                fishbone[x][0] = d
                break
            elif d > fishbone[x][1] and fishbone[x][2] == None:
                fishbone[x][2] = d
                break
        else:
            fishbone.append([None, d, None])

    return fishbone
```

```python
ans3dict = {k: fishboner2(v) for k, v in data3dict.items()}
```

```python
ans3list = sorted(ans3dict.items(), key= lambda b: [
    -int(''.join(str(f[1]) for f in b[1])),
    [-int(''.join(str(n) for n in f if n != None)) for f in b[1]],
    -b[0]
])
```

```python
sum(i * b[0] for i, b in enumerate(ans3list, 1))
```
