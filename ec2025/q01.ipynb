```python
from ecd import get_inputs
```

## Part 1

```python
data1 = get_inputs(event=2025, quest=1)['1']
```

```python
names1, directions1 = [d.split(',') for d in data1.split('\n\n')]
```

```python
clamp = lambda low, n, high: max(min(high, n), low)
```

```python
position1 = 0
for direction in directions1:
    match (direction[0], int(direction[1:])):
        case ('R', n):
            position1 = clamp(0, position1 + n, len(names1) - 1)
        case ('L', n):
            position1 = clamp(0, position1 - n, len(names1) - 1)
```

```python
print(names1[position1])
```

## Part 2

```python
data2 = get_inputs(event=2025, quest=1)['2']
```

```python
names2, directions2 = [d.split(',') for d in data2.split('\n\n')]
```

```python
position2 = 0
for direction in directions2:
    match (direction[0], int(direction[1:])):
        case ('R', n):
            position2 = (position2 + n) % len(names2)
        case ('L', n):
            position2 = (position2 - n) % len(names2)
```

```python
print(names2[position2])
```

## Part 3

```python
data3 = get_inputs(event=2025, quest=1)['3']
```

```python
names3, directions3 = [d.split(',') for d in data3.split('\n\n')]

for direction in directions3:
    match (direction[0], int(direction[1:])):
        case ('R', n):
            names3[0], names3[n % len(names3)] = names3[n % len(names3)], names3[0]
        case ('L', n):
            names3[0], names3[-n % len(names3)] = names3[-n % len(names3)], names3[0]
```

```python
print(names3[0])
```
