```python
from ecd import get_inputs

event, quest = 2025, 6
```

## Part 1

```python
data1 = get_inputs(event=event, quest=quest)['1']
tdata1 = "ABabACacBCbca"
```

```python
from collections import Counter
from string import ascii_lowercase, ascii_uppercase

mentors1 = Counter()
mentees1 = Counter()

for d in data1:
    if d in ascii_uppercase:
        mentors1[d] += 1
    elif d in ascii_lowercase:
        mentees1[d] += mentors1[d.upper()]

print(mentees1['a'])
```

## Part 2

```python
data2 = get_inputs(event=event, quest=quest)['2']
tdata2 = 'ABabACacBCbca'
```

```python
from collections import Counter
from string import ascii_lowercase, ascii_uppercase

mentors2 = Counter()
mentees2 = Counter()

for d in data2:
    if d in ascii_uppercase:
        mentors2[d] += 1
    elif d in ascii_lowercase:
        mentees2[d] += mentors2[d.upper()]

print(mentees2.total())
```

## Part 3

```python
data3 = get_inputs(event=event, quest=quest)['3']
tdata3 = 'AABCBABCABCabcabcABCCBAACBCa'
```

```python
from collections import Counter
from string import ascii_lowercase

mentees3 = Counter()

distance = 1000

truedata3 = data3 * 1000

for i, d in enumerate(truedata3):
    if d in ascii_lowercase:
        mentees3[d] += truedata3[max(0,i-distance):min(i+1+distance, len(truedata3))].count(d.upper())

print(mentees3.total())
```
