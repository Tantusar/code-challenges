## Part 3

```python
data3 = get_inputs(event=event, quest=quest)['3']

data3 = data3.split('\n')
```

```python
acc = [int(data3[0])]

for d in data3[1:-1]:
    acc.extend([int(n) for n in d.split('|')])

acc.extend([int(data3[-1])])

ans3 = 1

for x in range(0, len(acc), 2):
    ans3 *= acc[x] / acc[x+1]

ans3 * 100
```
