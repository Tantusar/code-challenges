```python
from ecd import get_inputs

event, quest = 2025, 7
```

## Part 1

```python
data1 = get_inputs(event=event, quest=quest)['1']

data1 = data1.split("\n\n")

names1 = data1[0].split(',')
followers1 = {}

for lt, _, fl in [d.partition(" > ") for d in data1[1].split('\n')]:
    followers1[lt] = fl.split(',')
```

```python
for nm in names1:
    for i in range(len(nm)-1):
        if not nm[i+1] in followers1[nm[i]]:
            break
    else:
        print(nm)
```

## Part 2

```python
data2 = get_inputs(event=event, quest=quest)['2']

data2 = data2.split("\n\n")

names2 = data2[0].split(',')
followers2 = {}

for lt, _, fl in [d.partition(" > ") for d in data2[1].split('\n')]:
    followers2[lt] = fl.split(',')
```

```python
sum2 = 0

for n, nm in enumerate(names2, 1):
    for i in range(len(nm)-1):
        if not nm[i+1] in followers2[nm[i]]:
            break
    else:
        print(nm)
        sum2 += n

print(sum2)
```

## Part 3

```python
data3 = get_inputs(event=event, quest=quest)['3']

data3 = data3.split("\n\n")

names3 = data3[0].split(',')
followers3 = {}

for lt, _, fl in [d.partition(" > ") for d in data3[1].split('\n')]:
    followers3[lt] = set(fl.split(','))
```

```python
working = set()
done = set()

for nm in names3:
    for i in range(len(nm)-1):
        if not nm[i+1] in followers3[nm[i]]:
            break
    else:
        working.add(nm)

while working:
    nm = working.pop()
    if nm in done:
        continue
    if len(nm) >= 7:
        done.add(nm)
    if len(nm) >= 11:
        continue
    
    for fl in followers3[nm[-1]]:
        working.add(nm + fl)

print(len(done))
```
