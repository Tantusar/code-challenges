```python
from ecd import get_inputs

event, quest = 2025, 10
```

## Part 1

```python
from collections import defaultdict

data1 = get_inputs(event=event, quest=quest)['1'].split('\n')

positions = defaultdict(set)

for x, i in enumerate(data1):
    for y, j in enumerate(i):
        if j != '.':
            positions[j].add((x, y))

walker = [
    ( 2,  1),
    ( 2, -1),
    (-2,  1),
    (-2, -1),
    ( 1,  2),
    (-1,  2),
    ( 1, -2),
    (-1, -2)
]

working = set()

for _ in range(4):
    for x, y in positions['D'].difference(working):
        working.add((x,y))

        for a, b in walker:
            positions['D'].add((x+a, y+b))

print(len(positions['D'].intersection(positions['S'])))
```

## Part 2

```python
from collections import defaultdict

data2 = get_inputs(event=event, quest=quest)['2'].split('\n')

tdata2 = '''...SSS##.....
.S#.##..S#SS.
..S.##.S#..S.
.#..#S##..SS.
..SSSS.#.S.#.
.##..SS.#S.#S
SS##.#D.S.#..
S.S..S..S###.
.##.S#.#....S
.SSS.#SS..##.
..#.##...S##.
.#...#.S#...S
SS...#.S.#S..'''.split('\n')

positions = defaultdict(set)

for x, i in enumerate(tdata2):
    for y, j in enumerate(i):
        if j != '.':
            positions[j].add((x, y))

walker = [
    ( 2,  1),
    ( 2, -1),
    (-2,  1),
    (-2, -1),
    ( 1,  2),
    (-1,  2),
    ( 1, -2),
    (-1, -2)
]

working = set()

thecount = 0

for _ in range(3):
    for x, y in positions['D'].difference(working):
        working.add((x,y))

        for a, b in walker:
            positions['D'].add((x+a, y+b))

    thecount += len(positions['D'].difference(positions['#']).intersection(positions['S']))

    positions['S'].difference_update(positions['D'].difference(positions['#']))

    positions['S'] = {
        (x+1, y) for x, y in positions['S'] if x < len(tdata2)
    }

    print(positions['S'])

    thecount += len(positions['D'].difference(positions['#']).intersection(positions['S']))

    positions['S'].difference_update(positions['D'].difference(positions['#']))

    print(thecount)

print(thecount)
```
