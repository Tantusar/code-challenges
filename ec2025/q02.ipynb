```python
from ecd import get_inputs

event, quest = 2025, 2
```

```python
from dataclasses import dataclass
from typing import Self

def neg_div(a: int, b: int) -> int:
    if a > 0:
        return a // b
    return (a + b - 1) // b


@dataclass(repr=False)
class Complex:
    x: int
    y: int

    def __add__(self, other: Self):
        return Complex(self.x + other.x, self.y + other.y)
    
    def __mul__(self, other: Self):
        return Complex(
            self.x * other.x - self.y * other.y, 
            self.x * other.y + self.y * other.x
            )
    
    def __truediv__(self, other: Self):
        return Complex(
            neg_div(self.x, other.x), 
            neg_div(self.y, other.y)
            )
    
    def __repr__(self):
        return f"[{self.x},{self.y}]"
```

## Part 1

```python
data1 = get_inputs(event=event, quest=quest)['1']

numbers1 = {}

for num in data1.split('\n'):
    var, _, val = num.partition('=')
    val = [int(v) for v in val.strip('[]').split(',')]

    numbers1[var] = Complex(*val)

numbers1['B'] = Complex(25, 9)
```

```python
ans1 = Complex(0, 0)

for x in range(3):
    ans1 *= ans1
    ans1 /= Complex(10, 10)
    ans1 += numbers1['A']
    
```

## Part 2

```python
data2 = get_inputs(event=event, quest=quest)['2']

numbers2 = {}

for num in data2.split('\n'):
    var, _, val = num.partition('=')
    val = [int(v) for v in val.strip('[]').split(',')]

    numbers2[var] = Complex(*val)

numbers2['B'] = Complex(35300, -64910)
```

```python
def checker2(val, verbose = False):
    if verbose:
        print(val)
        
    check = Complex(0, 0)

    for _ in range(100):
        check *= check
        if verbose: print(f"    {check}")
        check /= Complex(100_000,100_000)
        if verbose: print(f"    {check}")
        check += val

        if verbose:
            print(check)

        if (abs(check.x) > 1_000_000) or (abs(check.y) > 1_000_000):
            break
    else:
        return True
    
    return False
```

```python
print([checker2(Complex(x, y)) for x in range(numbers2['A'].x, numbers2['A'].x+1001, 10) for y in range(numbers2['A'].y, numbers2['A'].y+1001, 10)].count(True))
```
